```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
sch_data = "Resources/schools_complete.csv"
stu_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
sch_read = pd.read_csv(sch_data)
stu_read = pd.read_csv(stu_data)

# Combine the data into a single dataset.  
merg_data = pd.merge(stu_read, sch_read, how="left", on=["school_name", "school_name"])

#merg_data.head()
stu_read.head()
#sch_read.head(14)
```

## District Summary

* Calculate the total number of schools

* Calculate the percentage of students with a passing math score (70 or greater)

```python
Unq_schools = merg_data['school_name'].unique()
print(len(Unq_schools))

num_stu = sch_read['size'].sum()
print(num_stu)

tot_budget = sch_read['budget'].sum()
print(tot_budget)

avg_math = merg_data['math_score'].mean()
print(avg_math)

avg_read = merg_data['reading_score'].mean()
print(avg_read)
```

```python
math_score = merg_data.loc[merg_data['math_score'] >= 70]
num_PassMath = math_score.count()['student_name']
math_percPass = (num_PassMath / num_stu) * 100

read = merg_data.loc[merg_data['reading_score'] >= 70]
num_PassRead = read.count()['student_name']
read_percPass = (num_PassRead / num_stu) * 100


bothPass = merg_data[(merg_data['math_score'] >= 70) & (merg_data['reading_score'] >= 70)].count()['student_name']
both_percPass = (bothPass / num_stu) * 100

#print(both_percPass)
```

```python
#new table summarizing district data
summary = {"Total Schools": [len(Unq_schools)],
          "Total Students": [num_stu],
          "Total Budget": [tot_budget],
          "Average Math Score": [avg_math],
          "Average Reading Score": [avg_read],
          "Students Passing Math(%)": [math_percPass],
          "Students Passing Reading(%)": [read_percPass],
          "Students Passing Math and Reading(%)": [both_percPass]}

summary_df = pd.DataFrame(summary, columns = ["Total Schools",
          "Total Students",
          "Total Budget",
          "Average Math Score",
          "Average Reading Score",
          "Students Passing Math(%)",
          "Students Passing Reading(%)",
          "Students Passing Math and Reading(%)"])

summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,}".format)
summary_df.head(6)

```
